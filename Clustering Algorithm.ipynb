```python
# Create new dataframe with only continuous columns included.
dist_cluster = districts
dist_cluster = dist_cluster.drop(['District','County','District_type','Susp_Perc','Chronic_absent'], axis=1)
dist_cluster
```

```python
# Omit district outliers in terms of per pupil expenditure and enrollment.

dist_cluster['quintile_ppe'] = pd.qcut(dist_cluster['Per_pupil_exp'], 5, labels=np.arange(5,0,-1))
dist_cluster['quintile_enr'] = pd.qcut(dist_cluster['Enrollment'], 5, labels=np.arange(5,0,-1))
dist_cluster_filtered = dist_cluster[dist_cluster['quintile_ppe'] > 4]
dist_cluster_filtered = dist_cluster[dist_cluster['quintile_enr'] > 4]
```

```python
# Select 3 key features.
dist_cluster_filtered = dist_cluster_filtered[['FRL_Perc','ELA_metAbove','Per_pupil_exp']]
dist_plot = dist_cluster_filtered
# dist_plot = dist_plot.reset_index()
```

# Create Elbow Curve

```python
# Create an empty list to hold inertia values.
inertia = []
k = list(range(1, 11))
```

```python
# Loop through, looking for the best K value.
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(dist_cluster)
    inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
elbow_plot = df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
hvplot.show(elbow_plot)
```

```python
dist_cluster_filtered.hvplot.scatter(x="FRL_Perc", y="ELA_metAbove")
```

```python
# Function to cluster and plot dataset
def test_cluster_amount(dist_cluster, clusters):
    model = KMeans(n_clusters=clusters, random_state=5)
    model

    # Fitting model
    model.fit(dist_cluster)

    # Add a new class column to districts
    dist_cluster["class"] = model.labels_
```

```python
# Run the clustering function to show clusters.
test_cluster_amount(dist_cluster, 3)
dist_cluster.hvplot.scatter(x="FRL_Perc", y="ELA_metAbove", by="class")
```

```python
# 3D Plot.
fig = px.scatter_3d(
    dist_cluster,
    x="FRL_Perc",
    y="ELA_metAbove",
    z="Per_pupil_exp",
    color="class",
    symbol="class",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```
